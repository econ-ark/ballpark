### Summary
The paper provides a model in which households have identical and standard preferences and in which the tax system resembles the U.S. tax system, with the earnings process consistent with both the aggregate and the distribution of earinins of the U.S.

### The Model Economy
A modified version of the stochastic neoclassical growth model with uninsured idiosyncratic risk and no aggregate uncertianty.
* households face an uninsured, household-specific shock to their endownments of efficiency labor units
* households go through working and retiement stages in lify-cycles
* retired households face a positive probability of dying, and are replaced by a working-age descendant when they die
* households have bequest motive (altuistic toward their descendants)

#### A. The Private Sector
1. Population Dynamics and Information
    * a continuum of households, either in working or retirement age
        * For working people
            * faces idiosyncratic shock to endownment of efficiency labor
            * faces exogenous and positive probability of retiring
        * For retiring people
            * endowed with zero efficiency labor units
            * faces an exogenous and positive probability of dying
            * are replaced by a working-age descendant that inherits his estate when he dies

   One-dimensional shock $s$ is used to denote the random age and random effciency labor units endowment jointly. It is i.i.d. across household and follows a finite state Markov chain with conditional transition probabilities given by: 

$$\boldsymbol{\Gamma}_{S S}=\boldsymbol{\Gamma}\left(s^{\prime} \mid s\right)=\operatorname{Pr}\left\{s_{t+1}=s^{\prime} \mid s_{t}=s\right\}, \text { where } s \text { and } s^{\prime} \in S=\left\{1,2, \ldots, n_{s}\right\}.
$$

2. Employment Opportunities
    * every household is endowed with $\ell$ units of disposable time
    * joint age and endowment shock $ s \in S=\varepsilon \cup \mathcal{R}=\{1,2, \ldots, J\} \cup\{J+1,\ J+2, \ldots,\ 2J\}$
        * if $s \in \varepsilon$, he is of working age, and is endowed with $e(s)>0$
        * if $s \in \mathcal{R}$, he is retired, and is endowed with $e(s)=0$
            * $s \in \mathcal{R}$ is used to keep track of the realization of $s$ that the household faced during the last period of its working life
        * if $s \in \varepsilon, s' \in \mathcal{R}$, the household retires
        * if $s \in \mathcal{R}, s' \in \varepsilon$, the household has died and has been replaced by a working-age descendant.
3. Preferences

$$
E\left[\sum_{t=0}^{\infty} \beta^{t} u\left(c_{t}, \ell-l_{t}\right) \mid s_{0}\right]
$$
4. Production Possibilities

$$
\begin{aligned}
 Y_t &= f(K_t, L_t)\\
 Y_t&:\text{aggregate output}\\
 K_t&:\text{aggregate capital: sum of wealth of every household}\\
 L_t&:\text{aggregate labor input: sum of the efficiency labor units supplied by every household}\\
 \delta&: \text{constant capital depreciation rate}
\end{aligned}
$$

5. Transmission and Liquidation of Wealth
    * $z$ is the value of the household's wealth at the end of period $t$
    * every household inherits the estate of the previous member of its dynasty at the beginning of the first period of its working life
        * the inheritation happens after that the previous housihold's consumption and savings of last period have taken place
        * at the beginning of the following period, the descedant inherits $1-\tau_E(z_t)$ of the estate while the rest is transformed into the current-period consumption good taxed by the government
        
#### B. The Government Sector

The government taxes household's income and estates and uses them to make real transfers payment to retired households and to finance their consumption.
* government policy rule 
A specification of $\{\tau(y_t), \tau_E(z_t), \omega(s_t)\}$ and of a process on government consumption $\{G_t\}$
    * $\tau(y_t)$: income taxes, where $y_t$ denotes household income
    * $\tau_E(z_t)$: estate taxes
    * $\omega(s_t)$: public transfers
    * balanced budget
    $$
    \begin{aligned}
    G_t+Tr_t &= T_t\\
    Tr_t&:\text{aggregate transfers}\\
    T_t&:\text{affregate tax revenues}
    \end{aligned}$$
    
#### C. Market Arrangements
* no insurance and credit markets
* households can accumulate wealth in the form of real capital $a \in \text{compact set}\ \mathscr{A}$
* competitive spot production factor market

#### D. Equilibrium
Sate variables in the model is the 3-tuple $\{s_t, a_t, x_t\}$, where the aggregate state variable $x_t$ is defined over $\mathscr{B}$, an appropriate family of subsets of $\{S \times \mathscr{A}\}$. In steady state, we can only consider the pair $\{s_t, a_t\}$, and we can drop the time subscipt from all the current-period variables, and use primes to denote the values of variables one period ahead, due to ite recursive property.
##### 1. The Household's Decision Problem
$$
\begin{aligned}
v(s, a)&=\max _{c \geq 0, z \in \mathcal{L}, 0 \leq l \leq \ell}\left\{u(c, \ell-l)+\beta \sum_{s^{\prime} \in S} \mathbf{\Gamma}_{s s^{\prime}} v\left[s^{\prime}, a^{\prime}(z)\right]\right\}\\
\text{subject to}\ c+z &= y-\tau(y)+a\\
y &= a r+e(s) l w+\omega(s)\\
a^{\prime}(z) &= \left\{\begin{array}{ll}z-\tau_{E}(z) & \text { if } s \in \mathcal{R} \text { and } s^{\prime} \in \mathcal{E} \\
z & \text { otherwise }\end{array}\right.\\
r&: \text{rental price of capital}\\
w&: \text{wage rate}\\
y&: \text{income}\\
ar&: \text{capital income}\\
e(s)lw&: \text{labor income}\\
\end{aligned}
$$
The solution is $\{c(s,a), z(s,a), l(s,a)\}$, a set of functions that map the individual state into choices for consumption, gross savings, and hours worked.

##### 2. Definition of Equilibrium

A steady-state equilibrium consists of:
$$
\begin{aligned}
v(s,a)&: \text{household value function}\\
\{c(s,a), z(s,a), l(s,a)\}&: \text{household policy}\\
\{\tau(y), \tau_E(z), \omega(s), G\}&: \text{government policy}\\
\mathbf{x}&: \text{stationary probability measure of households}\\
(r,w)&: \text{factor prices}\\
\{K, L, T, T_r\}&: \text{macroeconomic aggregates}\\
\end{aligned}
$$
such that

(i) Factor inputs, tax revenues, and transfers are obtained aggregating over households:
$$
\begin{aligned}
K&=\int ad\mathbf{x}\\
L&=\int l(s, a) e(s) d \mathbf{x} \\
T&=\int \tau(y) d \mathbf{x}+\int \xi_{s \in R} \gamma_{s} \tau_{E}(z) z(s, a) d \mathbf{x} \\
T r&=\int \omega(s) d \mathbf{x}
\end{aligned}
$$
where $\varepsilon$ denotes the indicator function, $
\gamma_{s \varepsilon} \equiv \sum_{s^{\prime} \in \mathcal{E}} \boldsymbol{\Gamma}_{s, s^{\prime}}
$, and thus $\xi_{s \in \mathcal{R}} \mathcal{Y}_{s} \varepsilon$ is the probability that a type $s$ household dies, which is zero when $s\in S$. All integrals are defined over the state space $S \times \mathscr{A}$

(ii) Given $\mathbf{x}, K, L, r,$ and $w$, the household policy solves the households' decision problem, and factor prices are factor marginal productivities: 
$$
r = f_1(K,L)-\delta, w=f_2(K,L)
$$

(iii) The goods market clear:
$$
\int [c(s,a)+z(s,a)]d\mathbf{x} + G = f(K,L)+(1-\delta)K
$$
(iv) THe government budget constraints is satisfied:
$$
G+Tr = T
$$

(v) The measure of households is stationary:
$$
\mathbf{x}(B)=\int_{B}\left(\int_{S, \mathcal{U}}\left\{\xi_{z(s, a)} \xi_{s \notin R \vee s^{\prime} \notin \mathcal{E}}+\xi_{\left[1-\tau_{E}(z)\right] z(s, a)} \xi_{s \in \mathcal{R} \wedge s^{\prime} \in \mathcal{E}}\right\} \mathbf{\Gamma}_{s, s^{\prime}} d \mathbf{x}\right) d z d s^{\prime}
$$
for all $B\in\mathscr{B}$, where $\lor$ and $\land$ are the logical operators "or" and "and".

#### A. Functional Forms and Parameters

1. households' common utility function
$$
u(c, l) = \frac{c^{1-\sigma_1}}{1-\sigma_1}+\chi\frac{(\ell - l)^{1-\sigma_2}}{1-\sigma_2}
$$

2. The joint Age and Endowment of Efficiency Labor Units Process
$$
\boldsymbol{\Gamma}_{S S}=\left[\begin{array}{cc}
\boldsymbol{\Gamma}_{\varepsilon \varepsilon} & \boldsymbol{\Gamma}_{\varepsilon \mathcal{R}} \\
\boldsymbol{\Gamma}_{\mathcal{R} \varepsilon} & \boldsymbol{\Gamma}_{\mathcal{R} \mathcal{R}}
\end{array}\right]
$$

    $\boldsymbol{\Gamma}_{\varepsilon \varepsilon}$: the changes in the endowments of efficiency labor units of working-age households that are still of working age one period later; $\boldsymbol{\Gamma}_{\varepsilon \mathcal{R}}$: the transitions from the working-age states into the retirement states; $\boldsymbol{\Gamma}_{\mathcal{R} \varepsilon}$: the transitions from the retirement states into the working-age states that take place when a retired household dies and is replaced by its working-age descendant; $\boldsymbol{\Gamma}_{\mathcal{R} \mathcal{R}}$: changes in the retirement states of retired households that are still retired one period later
    $$
    \begin{aligned}
    \boldsymbol{\Gamma_{\varepsilon\varepsilon}}&:J \times J\text{matrix requiring}J^2\text{parameters}\\
    \boldsymbol{\Gamma_{\varepsilon\mathcal{R}}}&= p_{e\varrho}\mathbf{I}\\
    \boldsymbol{\Gamma_{\mathcal{R}\mathcal{R}}}&= p_{\varrho\varrho}\mathbf{I}\\
    \end{aligned}
    $$
see appendix for $\boldsymbol{\Gamma}_{\mathcal{R} \varepsilon}$ consists of $\phi_1, \phi_2,$ and $\gamma_{\varepsilon}^{*}$

### V. Findings
#### A. The Benchmark Model Economy as a Theory of Inequality

1. The endowment of efficiency labor units process

    Table 4 shows the four shocks are persistent. A large fraction of the differences in the economic performance of households may already have occurred before their members enter the labor market.

3. The life cycle profile of earnings and the intergenerational transmission of earnings ability.

4. Income taxes

    The income tax function is identical to the effective income tax function estimated by Gouveia and Strauss (1994).

    The distributions of earnings are very similar in both economies and the benchmark model economy does a significantly better job of accounting for the observed distribution of earnings than any of the previous attempts in the prvious literature. Authors conjecture that the differences between the Lorenz curves of earnings in the model economy and in the data would have been smaller if they had chosen a process on $s$ of a higher dimension.

8. The distribution of wealth

    We find that the benchmark model economy accounts for the U.S. distribution of wealth almost exactly and that it does a particularly good job of accounting for the top 1 percent of the distribution.
    
9. The distribution of consumption

    Consumption is more unequally distributed in the model economy than in the U.S. economy. The shares of the lowest four quintiles resemble the data significantly more than those of the top quintile. The distributions of consumption in the United States and in the model economies are almost the same after excluding the wealthiest 1 percent of the model economy.

### VI. A Policy Experiment: Abolishing Estate Taxation
We find that abolishing estate taxation brings about an increase in the steady-state output of 0.35 percent and an increase in the steadystate stock of capital of 0.87 percent in steady state.

### VII. Concluding Comments
The model shows that uninsured idiosyncratic earnings risk, retirement, altruism, and government transfers to retired households are essential ingredients to replicate the oberved earnings to wealth ratios of both rich and the poor households. Earnings process calibration is necessary to replicate the observed distributions of earnings and wealth sufficiently. No preference distinguishment between the poor and the rich is supposed in this model. Instead, two groups can be seen as the same type of people who have been subject to a different set of circumstances.
