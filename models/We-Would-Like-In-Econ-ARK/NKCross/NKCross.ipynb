### New Keynesian Cross: Bilbiie 2020

### Motivation:
The idea of the Keynesian Cross and with it the Fiscal Multiplier which is $\frac{1}{MPC}$ hinges on homogeneous agents. In case agents have different MPCs, the dynamics change.

Bilbiie (2020) creates an analytical framework how to transfer the idea of the Keynesian Cross into new keynesian models with representative and heterogeneous agents (RANK, TANK and HANK). As a result, the exercise affords analytical insights into monetary, fiscal, and forward guidance multipliers.

### New Keynesian Cross:

 
\begin{align*}
\text{PE curve: } C = C(Y,r) & \text{ with } 0 < C_Y < 1; C_r <0 \\
\omega & \text{ MPC: slope of the PE}\\
\Omega_D & \text{ autonomous expenditure change: shift of the PE}
\end{align*}

Hence the multiplier is
$$
\Omega = \frac{\Omega_D}{1-\omega}
$$

### NK cross in RANK

#### Proposition 1.
In RANK, the MPC $\omega$, autonomous expenditure increase $\Omega_D$ and multiplier $\Omega$ (for an interest rate cut of persistence $p$) are:
\begin{align*}
    &\omega^* = \frac{1 - \beta}{1 - \beta p}; & \Omega^*_D = \frac{\sigma \beta}{1 - \beta p}; & & \Omega^* = \frac{\sigma}{1 - p}
\end{align*}

### NK cross in TANK
See: Bilbiie, 2008
1. H: Hand-to-Mouth: mass of $\lambda$
2. S: Savers: mass of $1 - \lambda$
3. $\chi$: denotes the elasticity of H's consumption (and income) to aggregate income $y_t$
\begin{align}
    c_t^H = y_t^H = \chi y_t\\
    \chi = 1 + \phi\Big(1 - \frac{\tau^D}{\lambda}\Big)
\end{align}

#### Proposition 2.
In TANK, the MPC $\omega$, autonomous expenditure increase and multiplier $\Omega$ (for an interest rate cut of persistence $p$) are:
\begin{align*}
    &\omega^* = \frac{1 - \beta(1 - \lambda \chi)}{1 - \beta p(1 - \lambda \chi)}; & \Omega^* = \frac{\sigma}{1 - p} \frac{1 - \lambda}{1 - \lambda \chi}
\end{align*}
There is amplification $\Big(\frac{\partial \Omega}{\partial \lambda} > 0 \Big)$ if and only oif income inequality is countercyclical:
\begin{align*}
\chi > 1
\end{align*}


### NK in an analytical HANK model
See: Bilbiie, 2018
1. Exogeneous stochastic change of status between constrained H and unconstrained S (idiosyncratic uncertainty)
2. Insurnace is full within type (after idiosyncratic uncertainty is revealed), bu limited across types
3. Different asset liquidity: bonds are liquid (can be used to self-insure, before idiosyncratic uncertainty is revealed), while stocks are completely illiquid (cannot be used to self-insure)
4. No bond trading (no equilibrium liquidity)

Hence, mass of Hand-to-Mouth:
$$
\lambda = \frac{1 - s}{2 - s - h}
$$
And the multipliers are:
\begin{align*}
    &\omega = \frac{1 - \beta(1 - \lambda \chi)}{1 - \delta \beta p(1 - \lambda \chi)}; & \Omega = \frac{\sigma}{1 - \delta p} \frac{1 - \lambda}{1 - \lambda \chi}
\end{align*}


#### Additional Application: forward guidance puzzle
#### Proposition 4:
The multiplier of forward guidance FG (an interest rate cut in T periods) and the MPC in the analytical HANK model are:
\begin{align*}
    &\Omega^F_T = \sigma \frac{1 - \lambda}{1 - \lambda \chi}delta^T; & \omega^F_T = 1 - \Big(\beta(1 - \lambda \chi)\Big)^{1+T}
\end{align*}
The multiplier decreases with the horizon $\frac{\partial \Omega_T^F}{\partial T} < 0$, thus resolving the FG puzzle, if and only if there is discounting $\delta < 1$. While in the compounding case, the multiplier increases with the horizon $\frac{\partial \Omega_T^F}{\partial T} > 0$ and the FG puzzle is aggravated.
