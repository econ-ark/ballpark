## Financial Heterogeneity and the Investment Channel of Monetary Policy (2019)
## Pablo Ottonello and Thomas Winberry (R&R in Econometrica)

## The paper's objective


* Ottonello and Winberry study the role of financial frictions and firm heterogeneity in determining the
  investment channel of monetary policy. 

* Empirically, they find that firms with low default risk – those with low debt burdens, high credit ratings, and large "distance to default” – are the most responsive to monetary shocks using a
  heterogeneous firm New Keynesian model with default risk. 

* In their model, low-risk firms are more responsive to monetary shocks because their marginal cost of financing
  investment is relatively flat. The aggregate effect of monetary policy therefore depends on the distribution of default risk,   which varies over time. The results in essence, seem to follow the pattern of results from HA consumption models.

## Empirics

* First, they find that firms with low debt burdens, as measured by their leverage ratios, are significantly more responsive to monetary policy shocks than are high debt burdens. 

* Second, they argue that these heterogeneous responses are driven by heterogeneity in default risk.


# Data Desciption

**Monetary Policy Shocks** are combined with **firm-level outcomes** from the quarterly **Compustat** data. (Discrete time)

## Monetary Policy Shocks

$\varepsilon^{m}_{t}=\tau_{t}*(ffr_{t+\Delta_{+}}-ffr_{t-\Delta_{-}})$ (1)

They construct their shock $\varepsilon^{m}_{t}$ as expression 1 where t is the time of the monetary announcement, $ffr_{t}$
is the implied Fed Funds Rate from a current-month Federal Funds future contract at time t, $\Delta_{+}$ and $\Delta_{-}$ control the size of the time window around the announcement, and $\tau_{t}$ is an adjustment for the timing of the
announcement within the month. More details can be looked in the [paper](https://faculty.chicagobooth.edu/thomas.winberry/research/ottonello_winberry_monetary.pdf).

High-frequency shocks are aggregated to the quarterly frequency in order to merge them with their firm-level data.

## Firm Level Data

They draw firm-level variables from quarterly [Compustat](https://wrds-www.wharton.upenn.edu/pages/support/data-overview/wrds-overview-compustat-north-america-global-and-bank/) , a panel of publicly listed U.S. firms. Compustat satisfies three key requirements for their study: it is
quarterly, a high enough frequency to study monetary policy; it is a long panel, allowing
them to use within-firm variation; and it contains rich balance-sheet information, allowing them
to construct their key variables of interest. The main disadvantage of Compustat is that
it excludes privately held firms which are likely subject to more severe financial frictions. To correct for this, they calibrate the economic model   to match a broad sample of firms, not only the ones available in Compustat.

A summary statistic of firm level variables is provided above.

## Chief features of the Model

Standard New Keynesian Model

*  Model in three blocks in the spirit of Gali:
 * an investment block, which captures heterogenous responses to monetary policy;
    * The investment block contains a fixed mass of heterogeneous production firms that invest in capital subject to financial       frictions. It builds heavily on the flexible-price model developed in Khan, Senga and Thomas (2016).
      * First, they add idiosyncratic capital quality shocks, which helps them match observed default rates.
      * Second, they incorporate aggregate adjustment costs in order to generate time-variation in the relative price of  
        capital.
      * Third, they assume that new entrants have lower initial productivity than average firms, which helps them match   
        lifecycle dynamics.
 
 * a New Keynesian block, which generates a Phillips curve; 
 
    * Following Bernanke, Gertler and Gilchrist (1999), they keep the nominal rigidities separate from the investment block of        the model.
    
 * a representative household which closes the model.
 
    * All firms and household optimize
    * Lenders price default risk competitively
    * Distribution of firms is consistent with decision rules
    * All markets clear

## Calibration with Micro data

Some parameters are chosen exogenously and some are matched with moments in the data.

## Results

In this paper, the authors argue that financial frictions dampen the response of investment for
firms with high default risk. 

* First, the authors show in the micro data that firms with high leverage or low credit ratings invest significantly less than other firms following a monetary policy shock.
* Second, they build a heterogeneous firm New Keynesian model with default risk that is quantitatively consistent with these empirical results. In the model, monetary policy stimulates investment through a combination of direct and indirect effects. High-risk firms are less responsive to these changes because their marginal cost of investment finance is higher than for low-risk firms.
* Third, the aggregate effect of monetary policy is primarily driven by these low-risk firms, which suggests a novel form of state dependence: monetary policy is less powerful when default risk in the economy is greater.
