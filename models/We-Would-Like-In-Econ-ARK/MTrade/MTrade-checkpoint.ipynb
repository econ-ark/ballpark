### The Impact of Trade on Intra-Industry Reallocations and Aggregate Industry Productivity  
Published in Econometrica, Vol 71, No. 6. (Nov., 2003), pp. 1695-1725  
Author: Marc J. Melitz  
Notebook Creator: Jeongwon (John) Son

#### Summary  
The paper develops a dynamic model of firms with heterogeneous productivity levels to explain the effects of international trade on inter-firm reallocations of profit and market share. The more productive firms are motivated to enter the export market when exposed to trade which leads these firms to acuqire higher profit and market share. Less productive firms continue to produce only in the domestic market, while the least productive firms will be forced to leave when exposed to trade. Additionally, the reallocation of profit and market share generates aggregate industry productivity growth thereby contributing to a welfare gain, a channel which has not been theoretically examined before. Firms with different levels of productivity exist together in an industry because firms face sunk market entry costs while being intially uncertain of their productivity.

#### Bullet Point Summary  

- Firm heterogeneity is used to explain the effects of international trade exposure on intra-industry reallocation of profit and market share.
- Firms differ in their level of productivity.
- When exposed to international trade, more productive firms enter the export market and increase their profit and market share.
- When exposed to international trade, less productive firms do not enter the export market and continue producing only in the domestic market
- When exposed to international trade, the least productive firms leave the market.
- Aggregate industry productivity growth is generated by intra-industry/inter-firm reallocation, contributing to welfare gains.

#### The Model

#### Household Problem
- There is a representative consumer with a CES utility function over a continuum of goods indexed by $\omega$.

#### Firm Problem
- Firms are modeled within the framework of monopolistic competition
- Each firm chooses to produce a different variety $\omega$ and require only labor for production.
- The firm's profit, revenue and output all depend only on $\phi$, the productivity level of the firm

- Firms face an entry and exit choice.
- Prior to entry, firms are identical.
- To enter the industry, firms must pay a fixed entry cost $f_e >0$
- After entry, the productivity of firms $\phi$ is realized from a common distribution $g(\phi)$
- There is a threshold value of $\phi^*$ such that if $\phi < \phi^*$, then the firm will exit immediately after $\phi$ is realized.
- When exposed to international trade, firms can choose to pay a fixed cost $f_{ex} > 0$ to enter the export market. This occurs after productivity is realized.
- Trade provides the same opportunities to an open economy as would an increase in country size to a closed economy.
- There is a threshold value of ${\phi^*}_x$ such that if $\phi < {\phi^*}_x$, then the firm that decided to stay in the industry only produces for the domestic market.

#### Why is this paper important/interesting?

This paper presents a model that has become a foundation for explaining various effects/channels of trade through firm heterogeneity. It is also tractable since firm heterogeneity is integrated in such a way that the distribution of productivity levels for aggregate outcomes is representable by a sufficient statistic, an average firm productivity level. So, firms with productivity heterogeneity can be replaced with a representative firm.
