### **Summary**

This paper studies a heterogeneous-agent New Keynesian (HANK) model incorporating Diamond-Mortensen-Pissarides search and matching (SAM) from an analytical point of view. Due to the SAM structure, the model features an endogenous idiosyncratic earnings risk, which is empirically plausible to be countercyclical, and consequently, it induces a precautionary savings motive that results in downward pressure on real interest rates during recessions. In this setting, 

(a) the economy  may get stuck in a third type of steady state, a high-unemployment steady state,

(b) the monetary authority has to be more aggressive than the Taylor principle to eliminate indeterminacy, and

(c) price stickiness and the market incomepleteness (hence the endogenous earnings risks) work together to amplify the impact of aggregate shocks.

### **Non-Technical Overview**

By mathematically analyzing a model in which heterogeneous households have labor contract matching with firms, this paper examines the effects of increasing unemployment risk during recessions on monetary policy, unemployment, inflation, the effects of technological progress, and economic stability.

In conclusion, the increased risk of unemployment during this recession encourages households to make precautionary savings, which in turn reduces consumption and job offers, suggesting the possibility of a high-unemployment economy. It also suggests the existence of an economy with zero interest rates and positive inflation at the same time. It also suggests that in order to avoid economic instability, the central bank may need to conduct more aggressive expansionary monetary policy (interest rate cuts) in response to high unemployment.



### **The Model Features**

In this section, I will briefly describe the salient parts of the model.

#### **Households**

The household $i\in(0,1)$ maximizes the lifetime utility


\begin{equation}
\mathbb{E}_t \sum_{s=t}^{\infty} \beta^{s-t}\left(\frac{c_{i, s}^{1-\mu}-1}{1-\mu}-\zeta n_{i, s}\right)
\end{equation}

subject to the budget constraint

\begin{equation}
\begin{aligned}
c_{i, s} & +\frac{b_{i, s}}{R_s}+p_{h, s} h_{i, s} \leq z_i w_{i, s} n_{i, s}+\vartheta\left(1-n_{i, s}\right) +\frac{b_{i, s-1}}{\Pi_s}+\left(1-\tau_i\right)\left(p_{h, s}+d_{h, s}\right) h_{i, s-1}.
\end{aligned}
\end{equation}

with a borrowing constraint $b_{i, s} \geq-\psi z_i w_{i, s} n_{i, s}$. Here $\vartheta$ is the home production and $h_{i,s-1}$ is the equity holdings.


These constraints are simplified for each of the three different household types as follows.

(1) Employed workers ($z_i=1, n_{i,s}=1, \tau_i=1$) living on wage:

\begin{equation}
\begin{aligned}
c_{i, s} & +\frac{b_{i, s}}{R_s}+p_{h, s} h_{i, s} \leq w_{i, s}  +\frac{b_{i, s-1}}{\Pi_s}
\end{aligned}
\end{equation}

(2) Unemployed job searchers ($z_i=1, n_{i,s}=0, \tau_i=1$) living on home production:

\begin{equation}
\begin{aligned}
c_{i, s} & +\frac{b_{i, s}}{R_s}+p_{h, s} h_{i, s} \leq \vartheta\ +\frac{b_{i, s-1}}{\Pi_s}
\end{aligned}
\end{equation}


(3) Non-working capitalists ($z_i=0, n_{i,s}=0, \tau_i=0$) living on home production and capital income from investing in firm equity:

\begin{equation}
\begin{aligned}
c_{i, s} & +\frac{b_{i, s}}{R_s}+p_{h, s} h_{i, s} \leq \vartheta +\frac{b_{i, s-1}}{\Pi_s}+\left(p_{h, s}+d_{h, s}\right) h_{i, s-1}.
\end{aligned}
\end{equation}


#### **Firm-Worker Matching**

We introduce a standard DMP with Cob-Douglas matching function  $m_s=e_s^{\alpha} v_s^{1-\alpha}$ ($e$ for job searchers, $v$ for vacancies) so that we have

* Job-finding rate: $\eta_s=\theta_{s}^{1-\alpha}$

* Vacancy-filling rate: $q_s=\theta_s^{-\alpha}$,

where $\theta_s=\frac{v_s}{e_s}$ is the labor market tightness.

We have laws of motion

* Hours-worked dynamics: $n_s=(1-\omega)n_{s-1}+\eta_s e_s$

* Job searchers dynamics: $e_s=1-n_{s-1}-\xi+\omega n_{s-1}$,

where $\omega\in[0,1]$ is the job separation rate and $\xi \in(0,1)$ is the share of capitalists out of all households. 

The wage is determined via Nash bargaining, i.e., the wage solves 
\begin{equation}
\max \left(S_{i, s}^n\right)^v\left(S_s^f\right)^{1-v},
\end{equation}
where $S_{i, s}^n$ is the worker's match surplus and $S_S^f$ is the firm's surplus, and $v \in(0,1)$ is the worker's bargaining weight. We assume that were negotiations to fall through, the worker would become unemployed for at least one period and the firm could attempt to hire a new worker in the same period. Surpluses are given by

$S_{i,s}^n=V_{i,s}^n-V_{i,s}^u\hspace{20pt}$  ([Surplus for employed] = [Value of employment] - [Value of unemployment])

$S_{s}^f=\frac{\kappa}{q_s}\hspace{20pt}$ ([Surplus for matched firm] = [Hiring cost for 1 period]/[Vacancy-filling rate] = [Average hiring cost for 1 worker])


#### **Other Parts of the Model**

While I avoid explaining all the details of the model, I briefly describe the other parts of the model in words.

* Price setting: firms are monopolistically competitive and set the prices of their product to maximize the sum of discounted future profits as revenue minus input costs, a hiring cost for posting vacancies, and a Rotemberg (1982) price-adjustment cost.

* Monetary Policy: a Taylor rule with the ZLB.







### **Main Findings**


#### **Endogenous Earnings Risk Channel**


\begin{equation}
c_{n, s}^{-\mu}=\beta \mathbb{E}_s \frac{R_s}{\Pi_{s+1}} c_{n, s+1}^{-\mu}\left[1+\omega\left(1-\eta_{s+1}\right)\left(\left(\vartheta / w_{s+1}\right)^{-\mu}-1\right)\right]
\end{equation}

The Euler equation in this model differs from the one in the standard complete-markets model in that we have the **endogenous earnings risk wedge** (the term in squared brackets) that features the precautionary savings motive. 

*Storesletten, Telmer, and Yaron (2004)* study Panel Study of Income Dynamics (PSID) household income data and conclude that idiosyncratic labor market income risk is strongly *countercyclical*. 

Studying a 10% sample of all US working-age males, *Guvenen, Ozkan, and Song (2014)* show that the countercyclicality of earnings risk derives from increased left-skewness in recessions (i.e. higher likelihood of large earnings losses and lower likelihood of large gains) rather than from a countercylical variance.

The authors of HANK-SAM argue that the countercyclicality of earnings risk is more plausible based on an indirect check as well as a calibration of parameters. 

The countercyclical earnings risk implies a higher risk of earnings in recessions and hence a higher precautionary savings motive. This leads to a demand contraction and then in lower vacancy postings and wages., inducing a feedback mechanism.

#### **The Third Steady State**



![Steady-state Equilibria](Ravn_Sterk_2021_Fig2_Top_Left_Panel.png)

The model may have three steady states.

**I: Intended steady state**
* Corresponds to the steady state usually considered in NK literature
* In a HANK-SAM economy, the steady-state real interest rate is lower than the one in a complete market due to the precautionary motive

**II: Liquidity trap**
* A steady state with ZLB as in Benhabib, Schimidtt-Grohé, and Uribe (2001, 2002)
* In a HANK-SAM economy, a liquidity trap with a positive inflation rate is possible when the endogenous risk wedge is sufficiently countercyclical
* Precautionary motive reduces demand so much that the interest rate reaches zero even with $\pi>0$
* Purely expectational liquidity trap with $\pi>0$



**III: Unemployment trap**
* A steady state with a high unemployment rate, which does not appear in complete-market settings
* Can arise when the endogenous risk wedge is sufficiently countercyclical
* Expectations of poor labor market lead to a downward spiral towards an outcome where firms no longer want to hire
* An alternative (purely expectational) perspective of secular stagnation 

#### **Determinacy of the Dynamics**

**Determinacy around the intended steady state**

*A larger endogenous risk wedge unambiguously demands a more aggressive monetary policy to ensure
local determinacy of the intended equilibrium.*

The following other factors contribute to the determinacy of the HANK-SAM dynamics:

* Less price stickiness
* More competitive goods market
* More aggressive monetary policy against labor market tightness gap
* Larger labor adjustment cost
* More responsive wage to a change in job-finding rate

**Determinacy around the Unemployment Trap**

The unemployment trap is determinate under a standard Taylor rule, which responds
more than one-for-one to inflation.

#### **Productivity Shocks and Fluctuations**

* Analytically solves dynamics in response to a stationary AR(1) TFP shock
* Countercyclical endogenous earnings risk amplifies the impact of productivity shocks on the job-finding
rate relative to the acyclical risk case.
* Why? When the earnings risk is countercyclical, lower
productivity implies worse job-finding prospects, which stimulates employed workers’
desired savings, leading firms to post fewer vacancies, which induces worse job-finding
prospects, and so forth. 
* Higher price stickiness can amplify
the impact of productivity shocks on the job-finding rate when endogenous earnings risk is sufficiently countercyclical relative to the
stabilizing effects of monetary policy responses to labor market slack and wage
flexibility.
* This complementarity arises because the costlier it is to adjust prices,
the more hiring responds to variations in demand, and variations in hiring are the core of the amplification mechanism.
* In a HANK-SAM economy, inflation may increase in response to a positive productivity shock when the earnings risk wedge is sufficiently countercyclical because the demand stimulus from higher productivity may be so strong that the lower marginal cost effect is dominated.
* Authors find that empirically higher TFP either leaves inflation unchanged or gives rise to higher inflation.


#### **Monetary Policy Shocks and Fluctuations**
* Analytically solves dynamics in response to a stationary AR(1) MP shock
* An increase in the job-finding rate triggered by an expansionary MP shock is amplified by the presence of countercyclical earnings risk
* Why?  the market incompleteness will tend to provide a more
powerful role for monetary policy shocks in the case where the endogenous earnings
risk is countercyclical.
* More powerful role for monetary policy shocks

#### **Conclusion**

* (i) The HANK&SAM model features endogenous
earnings risk, which derives from an interaction between goods demand and labor
demand. This interaction is missing in NK models and in HANK models without
unemployment risk or other sources of endogenous idiosyncratic and uninsurable
income risk. 

* (ii) The endogenous earnings risk is countercylical when the downside
risk due to unemployment rising in recessions dominates the upside risk due to real
wages rising in booms. 

* (iii) When earnings risk is countercyclical, the impact of
productivity shocks and monetary policy shocks on job-finding rates and other real
variables tend to be amplified, relative to the standard NK model, or to models with
procyclical endogenous earnings risk. Moreover, in this case, nominal rigidities and
incomplete markets are complements. 

* (iv) The Taylor principle may fail to deliver
local determinacy of the intended steady state when earnings risk is countercyclical.
And, finally, 

* (v) The economy may get stuck in a high unemployment–low inflation
steady state, the unemployment trap, when endogenous earnings risk is sufficiently
countercyclical unless monetary policy is very aggressive.

#### **Limitations**

* While the model incorporated idiosyncratic risk and market incomletenesss, in equilibrium all households' bond holdings were set to zero; while this was an unavoidable assumption due to analytical tractability, it is at least potentially problematic when performing numerical calculations consistent with microdata.

* While the qualitative nature of empirically-plausible counter cyclical earnings risk has been clarified through analytical studies, numerical examination of how large the impact of this risk is is still awaited.
