# Prior Literature

{cite:t}`benhabib2019` builds on a long tradition of heterogeneous-agent macroeconomic models that attempt to explain the cross-sectional distribution of wealth. The foundational framework is the {cite:t}`bewley1983`–{cite:t}`huggett1993`–{cite:t}`aiyagari1994` class of incomplete-markets models, in which agents face idiosyncratic earnings risk and cannot fully insure against it, leading to precautionary saving and a nondegenerate wealth distribution. Within this framework, {cite:t}`castaneda2003` showed that a highly skewed and persistent earnings process could go some distance toward matching the US wealth distribution, and {cite:t}`krusell1998` demonstrated that stochastic discount factors could generate further dispersion. However, these earnings-driven models consistently struggled to reproduce the extreme concentration observed at the top of the wealth distribution — the thick right tail where the top 1% holds over a third of total wealth.

A second strand of literature identified additional mechanisms capable of generating thick tails. {cite:t}`denardi2004` introduced non-homogeneous bequest motives — the idea that the rich save a larger fraction of their wealth to leave bequests — and showed this could amplify concentration at the top. {cite:t}`quadrini2000` and {cite:t}`cagetti2006` emphasized the role of entrepreneurial risk, where heterogeneous returns to business investment generate wealth dispersion beyond what labor income alone can explain. {cite:t}`benhabib2011` provided the key theoretical result: in an economy with finitely-lived agents and stochastic idiosyncratic returns to wealth (capital income risk), the stationary wealth distribution follows a Pareto law in the tail, regardless of the earnings distribution. Empirically, {cite:t}`fagereng2020` documented from Norwegian tax records that returns to wealth are indeed heterogeneous and persistent across households, providing direct evidence for the capital income risk channel. {cite:t}`piketty2014` and {cite:t}`saez2016` further documented the extent and evolution of wealth concentration, motivating the quantitative exercise.

## Key Foundational Papers

- {cite:t}`aiyagari1994`: Established the canonical incomplete-markets model with idiosyncratic earnings risk and borrowing constraints, providing the workhorse framework for studying wealth inequality in general equilibrium.
- {cite:t}`benhabib2011`: Proved theoretically that stochastic idiosyncratic returns to wealth generate Pareto-tailed stationary wealth distributions — the direct theoretical precursor to the quantitative model in {cite:t}`benhabib2019`.
- {cite:t}`denardi2004`: Introduced wealth-dependent bequest motives (differential saving rates) as a mechanism for top wealth concentration, which {cite:t}`benhabib2019` operationalizes through the parameter μ < σ.
- {cite:t}`castaneda2003`: Showed how far a calibrated stochastic earnings process can go in matching the US wealth distribution within a Bewley-type model, providing a benchmark against which the additional mechanisms are measured.
- {cite:t}`fagereng2020`: Provided direct empirical evidence from Norwegian tax records that returns to wealth are heterogeneous and persistent across households, validating the key mechanism of stochastic idiosyncratic returns.

## Gap Addressed

Prior to {cite:t}`benhabib2019`, the three main channels that could generate thick-tailed wealth distributions — skewed earnings, differential saving/bequest rates, and stochastic returns to capital — had been studied largely in isolation. Standard Bewley/{cite:t}`aiyagari1994`/{cite:t}`huggett1993` models driven by earnings risk alone could not match the extreme concentration at the top of the US wealth distribution; models with bequest motives or entrepreneurial risk could match parts of the tail but did not jointly target the full distribution and intergenerational social mobility. No paper had embedded all three channels in a single parsimonious framework and then quantitatively measured their relative importance by matching both the wealth distribution (including the top 1% share) and the intergenerational mobility matrix simultaneously. {cite:t}`benhabib2019` closed this gap by combining all three mechanisms in a life-cycle model, estimating it via method of simulated moments, and showing that each factor plays a distinct and empirically identifiable role: capital income risk and differential savings drive the thick tail, while stochastic earnings are essential for generating realistic social mobility.
