# Subsequent Literature

Twenty-five papers published between 2021 and 2026 cite {cite:t}`benhabib2019`. The subsequent literature has expanded in three main directions.

**Deepening the return-heterogeneity channel.** {cite:t}`bruggemann2025` decompose entrepreneurial returns to wealth and their contribution to inequality, directly building on the finding that capital income risk is the primary driver of the thick right tail. {cite:t}`wolff2024` asks whether the rich earn higher returns specifically on housing — testing whether the mechanism operates through real estate. {cite:t}`best2023` introduce endogenous capital management, where wealthy households actively manage their portfolios in ways that amplify return heterogeneity. {cite:t}`benhabibcui2021` extend the framework to allow for rare large jumps in capital returns, showing these can further thicken the tail. On the empirical side, {cite:t}`snudden2025` estimates the joint distribution of idiosyncratic asset returns and wage risk for US households, providing updated microdata evidence on the magnitudes the model requires.

**Transitional dynamics and the time evolution of wealth concentration.** {cite:t}`benhabib2019` primarily worked under a stationarity assumption, with only a brief exploration of transition dynamics. {cite:t}`atkeson2022` study how rapidly top wealth shares change and the role of self-made fortunes versus inherited wealth, finding that top wealth dynamics are faster than standard models predict. {cite:t}`blanchet2022` develops econometric methods to uncover the dynamics of the full wealth distribution over time. {cite:t}`kaymak2026` quantitatively account for the historical rise in US wealth concentration, bringing policy changes (taxation, transfers) into the framework. These papers collectively push the field from matching a single cross-section toward explaining the full time path of inequality.

**Policy analysis and institutional variation.** {cite:t}`macnamara2023` structurally estimate a model with heterogeneous returns to study how marginal tax rates affect long-run income and wealth distributions. {cite:t}`cui2025` analyze optimal taxation of sudden capital income surges — a policy question that arises directly from the finding that stochastic returns drive tail inequality. {cite:t}`andersen2024` study how pension reforms interact with wealth inequality through differential savings. Meanwhile, several papers extend the theoretical foundations: {cite:t}`damato2024` study both tails of the distribution (not just the right tail), {cite:t}`lopezvelasco2025` develops an OLG model generating distinct power laws for earnings and wealth endogenously, and {cite:t}`birkner2022` analyze Pareto-distributed wealth dynamics in an open economy setting.

## Most Important Subsequent Papers

1. {cite:t}`atkeson2022`: Challenges the stationarity assumption by showing top wealth shares change faster than standard dynastic models predict, pointing to self-made fortunes and family firm dynamics as key missing ingredients.

2. {cite:t}`bruggemann2025`: Directly extends the return-heterogeneity channel by measuring how entrepreneurial returns differ from safe returns and quantifying their contribution to the wealth distribution’s thick tail.

3. {cite:t}`kaymak2026`: Moves beyond the stationary framework to explain the historical rise in US wealth concentration, incorporating policy changes in taxation and transfers.

4. {cite:t}`best2023`: Endogenizes the return heterogeneity that {cite:t}`benhabib2019` takes as exogenous, showing that wealthy households’ active portfolio management amplifies wealth inequality.

5. {cite:t}`benhabib2021`: Extends the framework by the same authors to study long-run intergenerational mobility with heterogeneous dynasties, deepening the mobility side of the distribution-and-mobility analysis.

## Open Questions

- **Endogenous return heterogeneity**: {cite:t}`benhabib2019` treats the stochastic return process as exogenous. How do portfolio choice, financial literacy, and access to investment opportunities endogenously generate return heterogeneity, and how does this interact with wealth dynamics?
- **Transitional dynamics and non-stationarity**: How well can these models explain the rapid changes in wealth concentration observed over recent decades (e.g., the post-1980 rise in the top 1% share)?
- **Within-lifetime risk**: The model assumes earnings profiles are deterministic within a generation. Incorporating within-lifetime earnings shocks, health risks, and mortality risk could substantially change implications for both the lower tail of wealth and social mobility.
- **Housing and heterogeneous assets**: The model has a single rate of return. How does the composition of wealth (housing, business equity, financial assets) interact with return heterogeneity to shape the distribution?
- **Policy design**: Given that capital income risk drives the tail, what are the optimal policy tools (wealth taxes, capital gains taxes, inheritance taxes) and how do they interact with the differential-savings and earnings channels?
